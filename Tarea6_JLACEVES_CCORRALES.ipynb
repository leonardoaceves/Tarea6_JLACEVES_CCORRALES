Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    
Demuestre matemáticamente que 
$$P(k+1)=\frac{\lambda}{k+1}P(k)$$
y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

Enlace: https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Poisson
    
$
\begin{aligned}
\frac{p_{k+1}}{p_k}& = \frac{e^{-\lambda}(\lambda)^k}{k!} \\
& = \frac{e^{-\lambda}(\lambda)^k}{k!}
\end{aligned}
$

Código de solución Jose Leonardo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def factorial(k):
    fact = 1
    for i in range(1,k+1):
        fact = fact * i
    return(fact)

def function(k,l):
    return((np.exp(-l) * l**k)/(factorial(k)))

#Parametros
k, k1, k2, k3 = np.random.randint(0,40), np.random.randint(0,40), np.random.randint(0,40), np.random.randint(0,40)
l = np.arange(0,30)
x = l
y = function(k, l)
y2 = function(k1, l)
y3 = function(k2, l)
y4 = function(k3, l)

#Ploteamos la función
#fig = figure(num=None, figsize=(6, 5), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.title('f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
